##### Code summary

This code performs a series of data cleaning and transformation steps in a DataFrame. Initially, it identifies and separates the columns into numeric and categorical. It then scales the numeric columns to ensure they all have the same magnitude. Uses the KNN algorithm to impute missing values ‚Äã‚Äãin scaled numeric columns, ensuring data is complete. Subsequently, it reverts the scale of the imputed data to the original form. The imputed numeric columns are then recombined with the categorical columns. Finally, it checks whether there are still missing values ‚Äã‚Äãand turns the 'Year' column into a datetime object to facilitate temporal analysis. These steps are performed to ensure the integrity and quality of the data, preparing it for further analysis.

```python
# Data Cleaning/Transforming

# Tranforming year in a datetime object
df['Year'] = pd.to_datetime(df['Year'], format= '%Y').dt.year

# Identify numeric and categorical columns
num_cols = df.select_dtypes(include=['float64', 'int64', 'int32']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Separate the columns
df_num = df[num_cols]
df_cat = df[cat_cols]

# Scale numeric columns
scaler = StandardScaler()
df_num_scaled = pd.DataFrame(scaler.fit_transform(df_num), columns=df_num.columns)

# Set the KNNImputer
imputer = KNNImputer(n_neighbors=5)

# Apply KNNImputer to scaled numeric columns
df_num_imputed_scaled = pd.DataFrame(imputer.fit_transform(df_num_scaled), columns=df_num.columns)

# Reverse the scale
df_num_imputed = pd.DataFrame(scaler.inverse_transform(df_num_imputed_scaled), columns=df_num.columns)

# Recombine the imputed numeric columns with the categorical columns
df_imputed_knn = pd.concat([df_cat.reset_index(drop=True), df_num_imputed.reset_index(drop=True)], axis=1)

# Check if there are still missing values ‚Äã‚Äãin the complete dataframe
print(df_imputed_knn.isna().sum())

```

##### Conclusion

All transformations and all null values ‚Äã‚Äãwere replaced successfully.

##### Code summary
In this code, the Body Mass Index (BMI) variable is categorized into five ranges: Underweight, Healthy Range, Overweight, Obesity and Severe Obesity. These categories are created using the Pandas pd.cut() method and added as a new column called 'BMI_Category' to the df_imputed_knn DataFrame. Life expectancy is then compared between these categories using a boxplot, which allows you to visualize the distribution of life expectancy within each BMI category. This type of analysis is useful for exploring the relationship between BMI and life expectancy and identifying possible patterns or differences between categories. Additionally, an ANOVA statistical test is performed to determine whether there are significant differences in life expectancy between BMI categories. The test result, including the test statistic and p-value, is printed to evaluate the statistical significance of the observed differences. This analysis is valuable to better understand the relationship between BMI and life expectancy and its health implications.

```python
# Creating BMI categories
bmi_categories = pd.cut(df_imputed_knn[' BMI '],
                        bins=[0, 18.5, 24.9, 29.9, 39.9, float('inf')],
                        labels=['Underweight', 'Healthy range', 'Overweight', 'Obesity', 'Severe obesity'])

# Add the new categories column to the dataframe
df_imputed_knn['BMI_Category'] = bmi_categories

# Creation of a comparative graph of life expectancy between BMI categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='BMI_Category', y='Life expectancy', data=df_imputed_knn, order=['Underweight', 'Healthy range', 'Overweight', 'Obesity', 'Severe obesity'])
plt.title('Life Expectancy by BMI category')
plt.xlabel('BMI category')
plt.ylabel('Life Expectancy')
plt.show()

# ANOVA statistical tests
underweight_life = df_imputed_knn[df_imputed_knn['BMI_Category'] == 'Underweight']['Life expectancy']
healthy_life = df_imputed_knn[df_imputed_knn['BMI_Category'] == 'Healthy range']['Life expectancy']
overweight_life = df_imputed_knn[df_imputed_knn['BMI_Category'] == 'Overweight']['Life expectancy']
obesity_life = df_imputed_knn[df_imputed_knn['BMI_Category'] == 'Obesity']['Life expectancy']
severe_obesity_life = df_imputed_knn[df_imputed_knn['BMI_Category'] == 'Severe obesity']['Life expectancy']

anova_result = f_oneway(underweight_life, healthy_life, overweight_life, obesity_life, severe_obesity_life)
print(f'ANOVA test between categories of BMI: statistic={anova_result.statistic}, p-value={anova_result.pvalue}')
```

##### The boxplot shows the distribution of life expectancy for each BMI category

- Underweight: The median life expectancy is around 70 years.
- Healthy range: The median life expectancy is lower, around 60 years.
- Overweight: The median is slightly above 60 years old.
- Obesity: The median life expectancy is around 65 years.
- Severe obesity: The median life expectancy is the highest, close to 70 years.

##### Interpretation of the ANOVA Test

The ANOVA test results are as follows:

- F-statistic: 440.15456517815403
- P-value: 3.123470973985787e-297

##### Meaning of Results

- High F-Statistic:
A high F statistic indicates that there is a large variability between the means of different groups (BMI categories) compared to the variability within groups. This suggests that the differences between category means are significant.

- Small p-value:
A p-value of 3.123470973985787e-297 is extremely small. This means that the probability of observing such large differences between BMI categories, if in reality there were no differences (null hypothesis), is almost non-existent.
Therefore, we reject the null hypothesis with a very high level of confidence, indicating that there is a statistically significant difference in life expectancy between the different BMI categories.

##### Conclusion
The results indicate that BMI has a significant impact on life expectancy, and the differences observed between BMI categories are statistically significant. In particular, the graph suggests that life expectancy varies between categories, with some categories (such as underweight and severe obesity) showing a higher median life expectancy compared to other categories (such as healthy range and overweight).

The `obter_regiao` function, decorated with `@lru_cache` to cache results and avoid repeating calls, checks whether the country's region is in the cache; otherwise, it tries to obtain this information from an public API. The `obter_regioes_async` function performs this operation asynchronously for multiple countries simultaneously, using a `ThreadPoolExecutor`. After obtaining the regions, the code updates the DataFrame with this information and identifies countries that do not yet have a region defined and displays them.

```python
# Set cache for known regions
known_regions = {
    'Bolivia (Plurinational State of)': 'Americas',
    'Iran (Islamic Republic of)': 'Asia',
    'Venezuela (Bolivarian Republic of)': 'Americas',
    'Micronesia (Federated States of)': 'Oceania'
}

# Function to obtain the region of a country
@lru_cache(maxsize=None)
def get_region(country):
    # Check if the region is in the known cache
    if country in known_regions:
        return known_regions[country]
    
    # Try to get the region via API
    try:
        response = requests.get(f"https://restcountries.com/v3.1/name/{country}")
        response.raise_for_status()
        data = response.json()
        region = data[0]['region'] if data else None
        return region
    except requests.exceptions.RequestException as e:
        print(f"Error getting data for {country}: {e}")
        return None
    except (KeyError, IndexError) as e:
        print(f"No region found for {country}")
        return None

# Function to get regions from multiple countries asynchronously
def get_regions_async(countries):
    with ThreadPoolExecutor(max_workers=5) as executor:
        results = list(executor.map(get_region, countries))
    return results

# Get regions for all countries asynchronously
df_imputed_knn['Region'] = get_regions_async(df_imputed_knn['Country'])

# Identify countries without region
countries_without_unique_regions = set(df_imputed_knn[df_imputed_knn['Region'].isnull()]['Country'])

# Show countries without region
if countries_without_unique_regions:
    print("Single countries without defined region:")
    for country in countries_without_unique_regions:
        print(country)
else:
    print("All countries have an identified region.")

# Display DataFrame with the obtained regions (including null values)
print("\nDataFrame with the obtained regions:")
print(df_imputed_knn[['Country', 'Region']].head())

# Check the number of countries without a region after filling in the dictionary
print("\nNumber of countries without region after completion:")
print(df_imputed_knn['Region'].isna().sum())
```

##### Conclusion

The code utilizes a local cache for known regions of certain countries and performs asynchronous calls to retrieve regions for other countries using an external API. It checks if regions are available in the cache before querying the API. After execution, all countries in the dataset have their regions identified, as indicated by the zero count of countries without a region post-processing. This approach demonstrates efficiency in obtaining and populating regional data for analysis purposes.

##### Code summary

This code performs a statistical analysis and visualization of data related to life expectancy versus alcohol consumption by region. First, it identifies the unique regions in the dataset and calculates the correlation between alcohol consumption and life expectancy for each of them using the Pearson coefficient. Regions are then ranked based on the absolute value of these correlations, highlighting those that are most strongly correlated. The code then generates multiple scatter plots, one for each region, where each point represents a country. The color of the dots varies according to the magnitude of the correlation, helping to visualize how alcohol consumption may be associated with life expectancy in different regions. Additionally, a single colormap (colorbar) is added to all scatterplots, indicating the correlation values. This analysis aims to better understand the relationships between these variables in different regional contexts.

```python
# --- Third Set of Plots: Life Expectancy vs Alcohol by Region ---

# Statistical analysis and visualization for categorical variable 'Region'
regions = df_imputed_knn['Region'].unique()

# Calculate correlations for each region and sort them
correlations = {}
for region in regions:
    region_data = df_imputed_knn[df_imputed_knn['Region'] == region]
    if len(region_data) > 1:  # Ensure there are enough data points for plotting
        r, _ = pearsonr(region_data['Alcohol'], region_data['Life expectancy'])
        correlations[region] = r

# Sort regions by absolute value of correlation, closest to -1 or 1
sorted_regions = sorted(correlations.keys(), key=lambda x: abs(correlations[x]), reverse=True)

# Redefine figsize for smaller scatter plots and larger color bar
fig, axes = plt.subplots(len(sorted_regions), 1, figsize=(4.5, 3 * len(sorted_regions)), sharex=True)
fig.suptitle('Life Expectancy vs Alcohol by Region', fontsize='xx-large', y=0.95, fontweight='bold')

norm = plt.Normalize(-1, 1)
cmap = cm.coolwarm_r  # Inverted colormap

for i, region in enumerate(sorted_regions):
    region_data = df_imputed_knn[df_imputed_knn['Region'] == region]
    ax = axes[i]
    color = annotate(region_data, x_col='Alcohol', y_col='Life expectancy', ax=ax, cmap=cmap, norm=norm)
    sns.scatterplot(x='Alcohol', y='Life expectancy', data=region_data, ax=ax, color=color)
    ax.set_title(f'Region: {region}')
    ax.set_ylabel('Life Expectancy')
    
    # Remove x-axis label from all but the last plot
    if i < len(sorted_regions) - 1:
        ax.set_xlabel('')
    else:
        ax.set_xlabel('Alcohol')

# Adjust the layout to make space for the colorbar
plt.tight_layout(rect=[0, 0, 0.95, 0.95])

# Add a single colorbar for all scatter plots
scamap = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
scamap.set_array([])
cbar = plt.colorbar(scamap, ax=axes, orientation='vertical', fraction=0.1, pad=0.04, aspect=70)
cbar.set_label('Correlation Coefficient')

plt.show()
```
